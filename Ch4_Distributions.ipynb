## 4.1 Normal Distribution

```python
from scipy import stats
```

```python
arr = stats.norm.rvs(size=1000)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
plt.hist(arr,bins=20)
plt.show()
```

```python
arr = stats.norm.rvs(loc=1100,scale=400,size=15000)
```

```python
plt.hist(arr,bins=100)
plt.show()
```

```python
arr.mean() ,arr.std()
```

```python
def z_score(val,arr):
    loc = arr.mean()
    scale = arr.std()
    return (val -loc)/scale
arr = stats.norm.rvs(size=1500,loc=1100,scale=200)
z_score(1300,arr) 
```

## 4.1.1 Finding tail area

```python
from scipy.stats import norm
norm.cdf(1300,loc=1100,scale=200)
```

```python
plt.hist(arr,bins=25)
plt.axvline(x = 900, color = 'r', label = '68%')
plt.axvline(x = 1300, color = 'r')
plt.axvline(x = 700, color = 'y', label = '95%')
plt.axvline(x = 1500, color = 'y')
plt.axvline(x = 500, color = 'black', label = '99.7%')
plt.axvline(x = 1700, color = 'black')
plt.legend()
plt.show()
```
