## 4.1 Normal Distribution

```python
from scipy import stats
```

```python
arr = stats.norm.rvs(size=1000)
```

```python
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')
plt.hist(arr,bins=20)
plt.show()
```

```python
arr = stats.norm.rvs(loc=1100,scale=400,size=15000)
```

```python
plt.hist(arr,bins=100)
plt.show()
```

```python
arr.mean() ,arr.std()
```

```python
def z_score(val,arr):
    loc = arr.mean()
    scale = arr.std()
    return (val -loc)/scale
arr = stats.norm.rvs(size=1500,loc=1100,scale=200)
z_score(1300,arr) 
```

```python
import numpy as  np
x = np.arange(-4,4,0.001)
y = stats.norm.pdf(x,loc=0,scale=1)
plt.plot(x,y)
plt.axvline(x = -1, color = 'r', label = '68%')
plt.axvline(x = 1, color = 'r')
plt.axvline(x = -2, color = 'y', label = '95%')
plt.axvline(x = 2, color = 'y')
plt.axvline(x = -3, color = 'black', label = '99.7%')
plt.axvline(x = 3, color = 'black')
plt.legend()
plt.fill_between(x,y)
```

```python
x = np.arange(-4,4,0.001)
y = stats.norm.pdf(x,loc=0,scale=1)
plt.plot(x,y)
plt.axvline(x = -1, label = '68%',linestyle='--')
plt.axvline(x = 1,linestyle='--')
plt.axvline(x = -2, color = 'y', label = '95%',linestyle='--')
plt.axvline(x = 2, color = 'y',linestyle='--')
plt.axvline(x = -3, color = 'red', label = '99.7%',linestyle='--')
plt.axvline(x = 3, color = 'red',linestyle='--')
plt.fill_between(x,y,where=(x<=1)&(x>-1) )
plt.fill_between(x,y,where=(x>1)|(x<-1),color='y' )
plt.fill_between(x,y,where=(x>2)|(x<-2),color='red' )
plt.arrow(x=0,y=.2,dx=.9,dy=0,color='black',shape='full',head_length=.1 ,head_width=0.01,linestyle='--',mouseover=True)
plt.arrow(x=0,y=.2,dx=-.9,dy=0,color='black',linestyle='--',shape='full',head_length=.1 ,head_width=0.01)
plt.text(s="68%",x=-.1,y=.21)
plt.arrow(x=0,y=.05,dx=1.9,dy=0,color='black',shape='full',head_length=.1 ,head_width=0.01,linestyle='--',mouseover=True)
plt.arrow(x=0,y=.05,dx=-1.9,dy=0,color='black',linestyle='--',shape='full',head_length=.1 ,head_width=0.01)
plt.text(s="95%",x=-.1,y=.06)
plt.arrow(x=0,y=.01,dx=2.9,dy=0,color='black',shape='full',head_length=.1 ,head_width=0.01,linestyle='--',mouseover=True)
plt.arrow(x=0,y=.01,dx=-2.9,dy=0,color='black',linestyle='--',shape='full',head_length=.1 ,head_width=0.01)
plt.text(s="99.7%",x=-.1,y=.02)
```

```python
from scipy.stats import geom
p = 0.5
mean,var,skew,kurt = geom.stats(p,moments='mvsk')
mean,var
```

```python
x = np.arange(geom.ppf(0.01,p),geom.ppf(0.99,p))
plt.plot(x,geom.pmf(x,p),'bo',ms=8, label='geom pmf')
plt.vlines(x, 0, geom.pmf(x, p), colors='b', lw=5, alpha=0.5)
```

```python
p=0.7
x = np.arange(geom.ppf(0.001,p),geom.ppf(0.999,p))
plt.plot(x,geom.pmf(x,p),'bo',ms=8, label='geom pmf')
plt.vlines(x, 0, geom.pmf(x, p), colors='b', lw=5, alpha=0.5)
```
