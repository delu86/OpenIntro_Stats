Transforming data: when data are very strongly skewed. we sometimes transform them so they are easier to model

```python
plt.hist(df_county['pop2017'],bins=10)
plt.ylabel('Frequency')
plt.xlabel('Population (m=millions)')
```

```python
df_county['pop2017_log'] = np.log10(df_county['pop2017'])
df_county.loc[:,['pop2017','pop2017_log']]
```

```python
plt.hist(df_county['pop2017_log'],bins=10)
plt.ylabel('Frequency')
plt.xlabel('log10(Population)')
```

## 2.2 Considering categorical data

```python
df_loans = pd.read_csv('csv/loans_full_schema.csv')
df_loans.head(10)
```

```python
df_loans_contingency = pd.crosstab(df_loans['application_type'],df_loans['homeownership'])
df_loans_contingency
```

```python
df_loans_contingency['Total'] = df_loans_contingency.sum(axis=1)
df_loans_contingency
```

```python
df_loans_contingency2 = df_loans_contingency.sum(axis=0)
df_loans_contingency2 = df_loans_contingency2.rename('Total')
df_loans_contingency = df_loans_contingency.append(df_loans_contingency2)
```

```python
df_loans_contingency_prop = pd.crosstab(df_loans['application_type'],df_loans['homeownership'],normalize='columns')
df_loans_contingency_prop
```

```python
df_loans_contingency_prop = pd.crosstab(df_loans['application_type'],df_loans['homeownership'],normalize='index')
df_loans_contingency_prop
```
